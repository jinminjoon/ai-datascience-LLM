### 분류분석

### 1.단순 감성분석

```python
from skllm.models.gpt.classification.zero_shot import ZeroShotGPTClassifier
import pandas as pd

reviews = [
    "이 영화는 정말 기쁘게 만들었다. 밝고 경쾌한 스토리와 멋진 연기가 돋보였다.",
    "엄청난 반전에 정말 놀랐다. 예상치 못한 결말이 인상적이었다.",
    "이 영화의 공포 분위기가 너무 무서워서 잠을 설쳤다.",
    "영화의 따뜻한 메시지와 깊은 감동이 마음을 울렸다.",
    "영화의 연출과 스토리가 너무 화가 날 정도로 실망스러웠다",
    "이 영화는 너무 감동적이여서 하루종일 울고 생각이 났다.",
    "배우들의 활기차고 밝은 연기가 기쁨을 주었다. 너무 즐거웠다.",
    "영화의 예상치 못한 전개에 놀라움을 금치 못했다.",
    "어두운 분위기와 소름 끼치는 장면들이 무서웠다.",
    "가족의 사랑과 희생에 대한 영화로, 매우 감동적이었다.",
    "주인공의 선택에 화가 나고 실망스러웠다. 기대 이하였다.",
    "재치 있는 대사와 상황이 유쾌한 웃음을 자아냈다."
]

clf = ZeroShotGPTClassifier()
clf.fit(None, ["기쁨","놀람","무서움","감동","화남","유쾌"])
labels = clf.predict(reviews)
```

```python
import pandas as pd
pd.DataFrame({"리뷰": reviews, "감성":labels})
```

### 2. 다중 감성 분석

```python
from skllm.models.gpt.classification.zero_shot import MultiLabelZeroShotGPTClassifier

mixed_emotion_reviews = [
    "이 영화는 기쁘면서도 동시에 놀라운 장면이 많았고, 감동적인 결말을 맞이했다.",
    "공포스러운 분위기 속에서도 유쾌한 요소가 있었으며, 놀라운 반전도 있었다.",
    "주인공의 행동에 화가 나면서도, 감동적인 배경이 이해가 되었다.",
    "영화가 너무 무서워서 긴장했지만, 감동적인 메시지가 있어서 기뻤다.",
    "유쾌한 코미디 장면들이 있었지만, 뜻밖의 슬픈 결말에 감동했다.",
    "놀라운 시각효과와 긴박한 스토리에 화가 났짐나, 결국에는 감동적이었다.",
    "영화의 재미있는 전개가 기쁨을 주었지만, 무서운 장면들이 많아서 불편했다.",
    "주인공의 모험에 놀라움을 느꼈지만, 결국 그의 실패에 화가 났다.",
    "영화의 유쾌한 대화가 웃음을 자아냈짐나, 갑작스런 슬픈 장면에 놀랐다.",
    "슬픈 장면이 많아 감동적이었짐나, 유쾌한 캐릭터들이 기쁨을 주었다.",
    "무서운 장면들이 있었지만, 주인공의 유쾌한 태도가 놀라웠다.",
    "영화의 감동적인 부분과 기쁜 순간들이 있었지만, 갑작스런 슬픔에 놀랐다."
]

emotion_labels = ["기쁨", "놀람", "무서움", "감동", "화남", "유쾌"]

clf = MultiLabelZeroShotGPTClassifier(model="gpt-4o", max_labels=3)
clf.fit(None, [emotion_labels])

labels = clf.predict(mixed_emotion_reviews)
```

```python
# 방법 1: 리스트 형태로 변환 (데이터 분석용으로 추천)
# labels가 numpy 배열이므로 .tolist()를 쓰면 [ ['기쁨'], ['슬픔', '놀람'] ... ] 형태로 바뀝니다.
df = pd.DataFrame({
    "리뷰": mixed_emotion_reviews,
    "감성": labels.tolist()
})

# 방법 2: 보기 좋게 문자열로 합치기 (보고서/출력용으로 추천)
# 예: ['기쁨', '놀람'] -> "기쁨, 놀람"
labels_str = [", ".join(label) for label in labels]

df_display = pd.DataFrame({
    "리뷰": mixed_emotion_reviews,
    "감성": labels_str
})

# 결과 확인
```

### 텍스트 요약

```python
from skllm.models.gpt.text2text.summarization import GPTSummarizer
from skllm.datasets import get_summarization_dataset

X = get_summarization_dataset()
print(X[0])
summarizer = GPTSummarizer(model='gpt-3.5-turbo', max_words=15)
X_summarized = summarizer.fit_transform(X)
print(X_summarized[0])
```

```python
for i in range(10):
  print(f"원문: {X[i]}")
  print("\n")
  print(f"요약문: {X_summarized[i]}")
  print("-"*50)
```

## 텍스트 요약

```python
from skllm.models.gpt.text2text.summarization import GPTSummarizer
from skllm.datasets import get_summarization_dataset

X = get_summarization_dataset()
print(X[0])
summarizer = GPTSummarizer(model="gpt-3.5-turbo", max_words=15)
X_summarized = summarizer.fit_transform(X)
print(X_summarized[0])
```

```python
for i in range(10):
  print(f"원문: {X[i]}")
  print("\n")
  print(f"요약문: {X_summarized[i]}")
  print("-"*50)
```

### 텍스트 번역

```python
from skllm.models.gpt.text2text.translation import GPTTranslator
from skllm.datasets import get_translation_dataset

X = get_translation_dataset()
t = GPTTranslator(model="gpt-3.5-turbo", output_language="korean")
translated_text = t.fit_transform(X)
```

```python
for i in range(6):
  print(f"원문: {X[i]}")
  print("\n")
  print(f"번역문: {translated_text[i]}")
  print("-"*50)
```
