```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')
```

```python
hi1 = model.encode("안녕하십니까?")
hi2 = model.encode("굿모닝입니다.")
hi3 = model.encode("나는 너무 힘들어")
```

```python
cosine_similarity([hi1], [hi2])
```

```python
cosine_similarity([hi1], [hi3])
```

```python
df = pd.read_csv('https://github.com/kairess/mental-health-chatbot/raw/master/wellness_dataset_original.csv')
df.head(7)
```

```python
df = df.drop(columns=['Unnamed: 3'])
df.head()
```

```python
df = df.dropna()
```

```python
df
```

```python
df.loc[1,'유저']
```

```python
model.encode(df.loc[1,'유저'])
```

```python
df['embedded'] = pd.Series([[]] * len(df)) # dummy
df['embedded'] = df['유저'].map(lambda x: list(model.encode(x)))
df.head()
```

```python
text = "요즘 친구가 도박에 빠져서 너무 힘들고 속을 썩이네"
embedding = model.encode(text)
embedding
```

```python
df['similarity'] = df['embedded'].map(lambda x: cosine_similarity([embedding], [x]).squeeze())
```

```python
df.head()
```

```python
df.loc[df['similarity'].idxmax()]
```

```python
ans = df.loc[df['similarity'].idxmax()]

print(ans['구분'])
print(ans['유저'])
print(ans['챗봇'])
print(ans['similarity'])
```

```python
import gradio as gr

def chatbot(user):
  embedding = model.encode(user)
  df['similarity'] = df['embedded'].map(lambda x: cosine_similarity([embedding], [x]).squeeze())
  answer = df.loc[df['similarity'].idxmax()]

  return answer['챗봇']

demo = gr.Interface(fn=chatbot, inputs='text', outputs='text')

demo.launch(debug=True, share=True)
```
